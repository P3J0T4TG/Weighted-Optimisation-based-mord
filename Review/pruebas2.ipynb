```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
###############
# LECTURA CSV #
###############

ruta="Ordinal/Results-CSSVC.csv"
dataset_list=['contact-lenses', 'pasture', 'squash-stored', 'squash-unstored', 'tae', 'newthyroid', 'balance-scale', 'SWD', 'car', 'bondrate', 'toy', 'eucalyptus', 'LEV', 'automobile', 'winequality-red', 'ESL', 'ERA']
df = pd.read_csv(ruta, header=[1])
df.head()
```

```python
######################
# ESQUELETO DF_FINAL #
######################

#creamos el df_final cuyas columnas seran dataset, MZE, MAE, Time 
df_final = pd.DataFrame(columns=["Dataset", "MZE", "MAE", "Time"])
#rellenamos la columna dataset con dataset_list
df_final["Dataset"] = dataset_list
df_final.head()
```

```python
# Calculamos la media de los valores de cada columna de df
df_mean = df.mean()
df_mean
```

```python
# rellenamos df_final con los elementos de df_mean, cada 3 filas de df_mean corresponden a un dataset
df_final["MZE"] = df_mean[0::3].values
df_final["MAE"] = df_mean[1::3].values
df_final["Time"] = df_mean[2::3].values
df_final.head()
```

```python
# Guardamos el df_final en un csv, de nombre usaremos la palabra que se encuentra despues de "Results-" en la ruta
nombre=ruta.split("Results-")[1]
nombre=nombre.split(".csv")[0]
df_final.to_csv("Ordinal/Resumen/"+nombre+".csv", index=False)
```
