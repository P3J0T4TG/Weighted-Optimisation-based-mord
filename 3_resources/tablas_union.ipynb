```python
import pandas as pd

def bold_min_in_row(df):
    """
    Reemplaza el valor mínimo de cada fila en un DataFrame con una cadena LaTeX en negrita,
    ignorando la primera fila y la primera columna.
    
    Args:
    df (pd.DataFrame): DataFrame de entrada con la primera fila y columna como palabras y el resto como números.
    
    Returns:
    pd.DataFrame: DataFrame con los valores mínimos de cada fila reemplazados por cadenas LaTeX en negrita.
    """
    def bold_min_latex(row):
        # Ignorar la primera columna al calcular el mínimo
        numeric_values = row[1:]
        min_val = numeric_values.min()
        return [row[0]] + [f"\\textbf{{{val}}}" if val == min_val else val for val in numeric_values]
    
    # Aplicar la función a partir de la segunda fila
    df.iloc[1:, 1:] = df.iloc[1:, 1:].apply(lambda row: pd.Series(bold_min_latex(row), index=row.index), axis=1)
    return df
```

```python
df_pom=pd.read_csv('Review/Ordinal/summary-svorimLin.csv')
df_=df_pom[['Dataset','MZEMean','MZESD']]
#seleccionamos los datasets presentes en list_datasets
df_=df_.loc[df_['Dataset'].isin(list_datasets)]
df_
```

```python
import pandas as pd

# Hacemos un DataFrame en blanco con las columnas 'Dataset' y 'POM'
df_final = pd.DataFrame(columns=['Dataset', 'SvorimLin'])

# Rellenamos la columna 'Dataset' con los datasets de list_datasets y en la columna 'POM' con los valores de AMAESMean +- AMAESSTD (en texto)
for dataset in list_datasets:
    pom_value = str(round(df_.loc[df_['Dataset'] == dataset]['MZEMean'].values[0],3)) + ' ± ' + str(round(df_.loc[df_['Dataset'] == dataset]['MZESD'].values[0],3))
    new_row = pd.DataFrame({'Dataset': [dataset], 'SvorimLin': [pom_value]})
    df_final = pd.concat([df_final, new_row], ignore_index=True)

#guardamos el csv
df_final.to_csv('Review/Ordinal/svorimLin_mze.csv', index=False)
# Mostrar el DataFrame final
print(df_final)
```
