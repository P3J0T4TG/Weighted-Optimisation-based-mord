```python
import pandas as pd
import numpy as np
```

```python
abalone = pd.read_csv('abalone.data',sep=" ", header=None)
abalone.head()
```

```python
#cambiamos los espacios por comas
abalone = abalone.replace(' ', ',', regex=True)

#guardamos el archivo
abalone.to_csv('abalone_mod.data', index=False, header=False)
```

```python
housing = pd.read_csv('housing.data', delim_whitespace=True, header=None)
housing.head()
```
