```python
import pandas as pd
import os
```

```python
# Cargamos los datasets
df_propio =pd.read_csv('Tabla_comparativa_MAE_(GridconCCR)(v2).csv')

df_review =pd.read_csv('./Review/Ordinal/Resumen/MAE_review.csv')
#le cambiamos el nombre a la columna Dataset
df_review.rename(columns={'Dataset':'dataset'}, inplace=True)

# Unimos los df que comparten la columna dataset
df = pd.merge(df_propio, df_review, on='dataset', how='inner')
df
```

```python
# Guardamos el df
df.to_csv('Tabla_comparativa_MAE_(PROPIO+REVIEW)_v2.csv', index=False)
```
